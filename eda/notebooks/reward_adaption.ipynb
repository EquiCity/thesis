```python
import numpy as np
import scipy.optimize
from matplotlib import pyplot as plt
```

```python
n = 600_000
t_max = np.log(n)
x = np.linspace(0,np.log(n),1000)
p_xs = np.array([0,t_max])
p_ys = np.array([100, 0.1])
func = lambda t,a,b: a*np.exp(-b*t)
popt = scipy.optimize.curve_fit(func,  p_xs,  p_ys,  p0=(t_max/2, 0.1))
```

```python
plt.figure()
plt.plot(p_xs, p_ys, 'ko', label="Original Noised Data")
plt.plot(x, func(x, *popt[0]), 'r-', label="Fitted Curve")
plt.legend()
plt.show()
```
