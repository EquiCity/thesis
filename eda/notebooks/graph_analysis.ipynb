```python
osm_graph = nx.read_gpickle('/home/rico/Documents/thesis/experiments/ams_data/test_graph/Amsterdam_osm_graph_2022-06-15.gpickle')

orig = (52.3525724975961, 5.00427352803761)
dest = (52.421329, 4.826383)

orig_node = ox.distance.nearest_nodes(osm_graph, 5.00427352803761, 52.3525724975961)
dest_node = ox.distance.nearest_nodes(osm_graph, 4.826383, 52.421329)
osmnx_route = ox.distance.shortest_path(osm_graph, orig_node, dest_node, 'length', cpus=None)


edge_lengths = ox.utils_graph.get_route_edge_attributes(osm_graph, osmnx_route, 'length')
route_len_m = sum(edge_lengths)
```

```python
import matplotlib.pyplot as plt

fig, ax = ox.plot_graph_route(
    osm_graph, osmnx_route, route_color='darkorange', show=False, close=False)

ax.scatter(
    osm_graph.nodes[orig_node]['x'], osm_graph.nodes[orig_node]['y'],
    c='lime', s=100, label='orig')

ax.scatter(
    osm_graph.nodes[dest_node]['x'], osm_graph.nodes[dest_node]['y'],
    c='red', s=100, label='dest')

ax.scatter(
    orig[1], orig[0],
    color='lime', marker='x', s=100, label='orig-point')

ax.scatter(
    dest[1], dest[0],
    color='red', marker='x', s=100, label='dest-point')

plt.legend()
plt.show()
```
